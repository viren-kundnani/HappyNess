## Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
!pip install keras
!pip install tensorflow
```

#### Getting data

```python
df = pd.read_csv('Dataset used for the main training.csv')
df.head()
```

```python
df = df.drop(['Positive affect','Negative affect'],axis = 1)
```

```python
df = df[df.year == 2020]
df
```

```python
df = df.dropna(axis=0)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.describe()
```

### Classifying the countries as happy/not happy

```python
df2 = pd.read_csv('2020 rankings used for classification.csv')
df2.head()
```

```python
df2['Country name'] = df2['Country name'].str.replace('[+]', '',regex = False)
df2
```

```python
df2 = df2.dropna()
```

```python
df = df.reset_index(drop = True)
```

```python
df['class'] = df2['Class']
df
```

```python
print(df['class'].value_counts())

label_encode = {"class": {"Happy":0,"Not Happy":1}}

df.replace(label_encode,inplace=True)

print(df['class'].value_counts())

y_values = df['class']

y_values = to_categorical(y_values)

print(y_values)
```

# Using K- Neighbors Classifier

```python
KNN = KNeighborsClassifier()

x = df[['Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']]
y = df['class']


KNN = KNN.fit(x,y)

test = pd.DataFrame()
test['Log GDP per capita'] = [8.830280304]
test['Social support'] = [0.638051987]
test['Healthy life expectancy at birth'] = [60.09999847]
test['Freedom to make life choices'] = [0.890443385]
test['Generosity'] = [0.071806304]
test['Perceptions of corruption'] =  [0.80526334] 
predict_happiness = KNN.predict(test)

if predict_happiness == [1]:
    print('The country is not happy')
else:
    print('The country is happy')
```

# Using neural network

## Standardising the data

```python
x_values = df[['Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']]
print(x_values.head())
standardise = StandardScaler()

x_values = standardise.fit_transform(x_values)
x_values_df = pd.DataFrame(x_values)
```

```python
print(x_values_df.head())
```

```python
print(x_values_df.describe())
```

## Making the model

```python
model = Sequential()

model.add(Dense(6,input_dim=6,activation='relu'))


model.add(Dense(6,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(2,activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

### Encoding

```python
print(df['class'].value_counts())

label_encode = {"class": {"Happy":0, "Not Happy":1}}

df.replace(label_encode,inplace=True)

print(df['class'].value_counts())

y_values = df['class']

y_values = to_categorical(y_values)

print(y_values)
```

## Training the neural network

```python
model.fit(x_values,y_values,epochs=20,shuffle=True)
```

```python
from sklearn.model_selection import train_test_split

x_values = df[['Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']]

x_train, x_test, y_train, y_test = train_test_split(x_values,y_values,test_size=0.25,random_state=10)
print("Number of rows in x_train:", x_train.shape[0])
print("Number of rows in x_test:", x_test.shape[0])
print("Number of rows in y_train:", y_train.shape[0])
print("Number of rows in y_test:", y_test.shape[0])
standardise = StandardScaler()
x_train = standardise.fit_transform(x_train)
x_test = standardise.transform(x_test)
```

```python
model_val = Sequential()
model_val.add(Dense(6,input_dim=6,activation='relu'))
model_val.add(Dense(6,activation='relu'))
model_val.add(Dense(2,activation='softmax'))
model_val.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model_val.fit(x_values,y_values,epochs=50,shuffle=True)
```

```python
model_val.fit(x_train,y_train,epochs=50,shuffle=True,validation_data=(x_test,y_test))
```

## Testing the model with extra data

```python
df3 = pd.read_csv('Extra data for testing.csv')
df3
```

```python
x_new = df3[['Log GDP per capita','Social support','Healthy life expectancy at birth','Freedom to make life choices','Generosity','Perceptions of corruption']]
x_new_scale = standardise.transform(x_new)
y_new = model_val.predict(x_new_scale)
```

```python
cntry = input('Enter the name of the country: ')
LGPC = float(input("Enter the Log GDP per capita: "))
SS = float(input('Enter the social support: '))
HLEAB = float(input('Enter the healthy life expectancy at birth: '))
FTMLC = float(input('Enter the freedom to make life choices: '))
G = float(input('Enter the generosity: '))
POC = float(input('Enter the perceptions of corruption: '))

KNN = KNeighborsClassifier()

x = df[['Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']]
y = df['class']


KNN = KNN.fit(x,y)

test = pd.DataFrame()
test['Log GDP per capita'] = [LGPC]
test['Social support'] = [SS]
test['Healthy life expectancy at birth'] = [HLEAB]
test['Freedom to make life choices'] = [FTMLC]
test['Generosity'] = [G]
test['Perceptions of corruption'] =  [POC] 
predict_happiness = KNN.predict(test)

if predict_happiness == [1]:
    print(cntry + ' ' + 'is not happy')
else:
    print(cntry + ' ' + 'is happy')
```
